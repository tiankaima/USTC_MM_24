```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read ./data/WHO-COVID-19-global-data.csv
df = pd.read_csv('./data/WHO-COVID-19-global-data.csv')
print(df.head())
```

```python
# plot CN data:
df_cn = df[df['Country'] == 'China']

# trim the data, < 2021-01-01:
# df_cn = df_cn[df_cn['Date_reported'] < '2020-04-10']
# df_cn = df_cn[df_cn['Date_reported'] > '2022-12-01']
# df_cn = df_cn[df_cn['Date_reported'] < '2023-02-01']


# print(df_cn.head())
# plot cumulative cases, deaths:
plt.figure()
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(df_cn['Date_reported'], df_cn['Cumulative_cases'], 'b-')
ax2.plot(df_cn['Date_reported'], df_cn['Cumulative_deaths'], 'r-')
ax1.set_xlabel('Date')
ax1.set_ylabel('Cumulative_cases', color='b')
ax2.set_ylabel('Cumulative_deaths', color='r')
ax1.xaxis.set_major_locator(plt.MultipleLocator(100))
plt.title('Cumulative cases and deaths in China')
plt.show()
plt.savefig('./output/COVID-19-CN.png')
```

```python
# plot CN data:
df_cn = df[df['Country'] == 'China']

# trim the data, < 2021-01-01:
# df_cn = df_cn[df_cn['Date_reported'] > '2023-02-01']
df_cn = df_cn[df_cn['Date_reported'] < '2024-01-01']


# print(df_cn.head())
# plot cumulative cases, deaths:
plt.figure()
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(df_cn['Date_reported'], df_cn['New_cases'], 'b-')
ax2.plot(df_cn['Date_reported'], df_cn['New_deaths'], 'r-')
ax1.set_xlabel('Date')
ax1.set_ylabel('New_cases', color='b')
ax2.set_ylabel('New_deaths', color='r')
ax1.xaxis.set_major_locator(plt.MultipleLocator(100))
plt.title('New cases and deaths in China')
# plt.show()
plt.savefig('./output/COVID-19-CN.jpg')
```

```python
# plot CN data:
df_cn = df[df['Country'] == 'China']

df_cn = df_cn[df_cn['Date_reported'] < '2020-04-10']

plt.figure()
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.scatter(df_cn['Date_reported'], df_cn['Cumulative_cases'], color='b')
ax2.scatter(df_cn['Date_reported'], df_cn['Cumulative_deaths'], color='r')
ax1.set_xlabel('Date')
ax1.set_ylabel('Cumulative_cases', color='b')
ax2.set_ylabel('Cumulative_deaths', color='r')
ax1.xaxis.set_major_locator(plt.MultipleLocator(100))
plt.title('Cumulative cases and deaths in China < 2020-04-10')
plt.savefig('./output/COVID-19-CN-20200410.jpg')
plt.show()
```
